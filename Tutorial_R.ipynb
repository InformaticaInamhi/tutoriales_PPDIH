Esta guía permite, instalar librerías de R, dentro del usuario de juputer mediante la conola de R on en un cuarderno jupyter de R, estos comandos permitiran instalar cualquier paquere dentro del entorno del cada usuario para evitar conflictos entre versiones de librerías.


## importante!!

## 1. Crear la carpeta para la librerías de usuario mediante el comando

```R
dir.create(Sys.getenv("R_LIBS_USER"), recursive = TRUE, showWarnings = FALSE)
```
Eso permitira crear un directoria para cada usuario en donde se instalara cada librería.

verificar que se creo la carpeta con el comando:
```R
Sys.getenv("R_LIBS_USER")
```
debe imprimir algo similar a lo siguiente:

```R
'/home/jupyter-drosero/R/x86_64-pc-linux-gnu-library/4.5'
```
Donde jupyter-**drosero** es el nobre de su usuario 

**! importante** se debe reiniciar el kernel de R en el menu **|Kernel|** **|Restart Kernel and clear Outputs of All Cells|**

## 2. Instalar las librerías.
la instalación se realiza mediante los comandos normales, únicamente se debe agregar lo opcion de donde se instalara la librería en este caso medinte el parametro **lib = Sys.getenv("R_LIBS_USER")** , como se muestra en el sigueinte ejemplo.



**Ejemplo 1**
```R
install.packages("pkgload", dependencies = TRUE, lib = Sys.getenv("R_LIBS_USER"))

```
**Ejemplo 2**
```R
install.packages("devtools", dependencies = TRUE, repos = "https://cloud.r-project.org", lib = Sys.getenv("R_LIBS_USER"))

```
